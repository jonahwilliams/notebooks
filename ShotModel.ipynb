```python
%matplotlib inline
import pandas as pd
import numpy as np
from ggplot import *
from __future__ import division
```

```python
data = pd.read_csv('C:/Users/Jonah/Dropbox/soccer project/DATA/SHOTS DATA/ShotsData.csv')
```

```python
data.head()
```

```python
data = data.fillna(0)
```

```python
data.head()
```

```python
data = data.drop(['assist_x','assist_y'], axis=1)
```

```python
data['SHOT_SUCCESS'] = data['event_name'].map(lambda x: (x == 'SAVED') or (x == 'GOAL')).astype(int)
data['SAVE_SUCCESS'] = data['event_name'].map(lambda x: (x == 'SAVED')).astype(int)
```

```python
for item in data.columns:
    print item
```

```python
ggplot(aes(x='x',y='y',color='SHOT_SUCCESS'),data=data) + \
    geom_point() + \
    xlim(low=0, high=100) + \
    ylim(low=0, high=100)
```

```python
ggplot(aes(x='x',y='y',color='SAVE_SUCCESS'),data=data) + \
    geom_point() + \
    xlim(low=0, high=100) + \
    ylim(low=0, high=100)
```

```python
ggplot(aes(x='x',y='y',color='SAVE_SUCCESS', group='team_id'),data=data[(data.team_id == 1) | (data.team_id == 3)]) + \
    geom_point() + \
    xlim(low=0, high=100) + \
    ylim(low=0, high=100) + \
    facet_wrap(x='team_id', ncol=1)
```

```python
ggplot(aes(x='x',y='y',color='SHOT_SUCCESS', group='214 Big Chance'),data=data) + \
    geom_point() + \
    xlim(low=0, high=100) + \
    ylim(low=0, high=100) + \
    facet_wrap(x='214 Big Chance')
```

```python
ggplot(aes(x='x',y='y',color='SHOT_SUCCESS', group='154 Intentional Assist'),data=data) + \
    geom_point() + \
    xlim(low=0, high=100) + \
    ylim(low=0, high=100) + \
    facet_wrap(x='154 Intentional Assist')
```

```python
data['IN_POSTS'] = data['y'].map(lambda x: 45 <= x <= 55)
```

```python
ggplot(aes(x='x',y='y',group='IN_POSTS', color='SHOT_SUCCESS'), data=data) + \
geom_point() + \
facet_wrap(x='IN_POSTS') + \
xlim(low=0, high=100) + \
ylim(low=0, high=100)
```

```python
data['DIST'] = np.sqrt(np.square(data['x'] - 100) + np.square(data['y'] - 50))
```

```python
ggplot(aes(x='DIST',group='SHOT_SUCCESS'), data=data) + geom_histogram() + facet_wrap('SHOT_SUCCESS')
```

```python
data['ANGLE'] = np.arccos((100 - data['x']) / data['DIST'])
```

```python
ggplot(aes(x='ANGLE', group='SHOT_SUCCESS'), data=data) + geom_histogram() + facet_wrap('SHOT_SUCCESS') 
```
