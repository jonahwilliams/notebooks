<h1>Splines and Stuff</h1>
<p>Piecewise cubic polynomials and natural cubic splines are an example of basis expansions.  in R, they can be used with</p>
  
    library(splines)
    ns(x, df=degrees)
    
<p>A piecewise cubic polynomial with 2 knots is a basis expansion of the form</p>
$h_1(X) = 1, \ h_2(X) = X, \ h_3 = X^2, \ h_4 = X^3, \ h_5 = (X - \xi_1)^3_+, \ h_6 = (X - \xi_2)^3_+$

According to the authors of ESL, there is seldom any reason to go beyond cubic splines.  Generally, the placements of the knots are specified by percentile distribution of X.

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('/Users/jonahwilliams/Documents/SAheart.data')
df = df.drop('row.names',axis=1)
df.famhist = df.famhist.map({'Present':1,'Absent':0})
df.head()
```

```python
glm = LogisticRegression()
X = df.tobacco.values
X.shape = (462, 1)
y = df.chd.values
glm.fit(X, y)
domain = np.linspace(0,30,300)
domain.shape = (300, 1)
pred = glm.predict_proba(domain)


sns.set_style('darkgrid')
plt.figure(figsize=(8, 6))
plt.scatter(df.tobacco, df.chd, color="black", alpha=0.5)
plt.plot(domain, pred[0::, 1])
plt.xlim(df.tobacco.min(), df.tobacco.max())
plt.title("Tobacco Usage versus Heart Disease")
plt.yticks([0, 1])
plt.show()
```

<p>First we'll consider a piecewise cubic polynomial with two splines.  the knots should be placed at the 0.33 and 0.66 percentiles</p>

```python
def makeSplines(data, knots=[]):
    n = len(data)
    expansion = np.zeros((n, 6))
    expansion[0::, 0] = 1
    expansion[0::, 1] = data
    expansion[0::, 2] = data**2
    expansion[0::, 3] = data**3
    if not knots:
        expansion[0::, 4] = (data - np.percentile(data, 33))**3
        expansion[0::, 5] = (data - np.percentile(data, 66))**3
    else:
        expansion[0::, 4] = (data - knots[0])**3
        expansion[0::, 5] = (data - knots[1])**3
    return expansion

expandedData = makeSplines(df.tobacco)
```

```python
glm = LogisticRegression()
X = expandedData
y = df.chd.values
glm.fit(X, y)
pred = glm.predict_proba(X)


sns.set_style('darkgrid')
plt.figure(figsize=(8, 6))
plt.scatter(df.tobacco, df.chd, color="black", alpha=0.5)
plt.xlim(df.tobacco.min(), df.tobacco.max())
plt.scatter(X[0::, 1], pred[0::, 1])
plt.title("Tobacco Usage versus Heart Disease")
plt.show()
```
