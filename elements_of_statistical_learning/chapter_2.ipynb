<h1>Overview of Supervised Learning</h1>
Set of variables denoted as inputs, which are measured/preset.  These have some influence over one or more outputs.  For each example the goal is to predict the value of the outputs.

qualitative data come from some finite set, such as the iris data.  predictions invvloving them are refered to as classifications.
quantitative data is some measured value.  For historical reasons (Galton) predictions involving them are referd to as regressions.

In terms of usefullness, I think of the following models as being the absolute minimum; useful as a standard to compare other models to but not much in of themselves.

```python
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
current_palette = sns.color_palette()
sns.set_style('darkgrid')
def color(x):
    return current_palette[2] if x > 0.5 else current_palette[0]
```

```python
# Artifical Dataset - Mixture of Gaussians
red = np.array([[1, 1, np.random.normal(2, 1), np.random.normal(2, 1)] for i in xrange(100)])
blue = np.array([[0, 1, np.random.normal(1, 1), np.random.normal(1, 1)] for i in xrange(100)])
plt.scatter(red[0::, 2], red[0::, 3], color=current_palette[2], label="Class 1")
plt.scatter(blue[0::, 2], blue[0::, 3], color=current_palette[0], label="Class 0")
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.ylabel("X2")
plt.xlabel("X1")
```

<h2>Least Squares and Nearest Neighbors</h2>
The linear model method makes huge assumptions about structure but yields stable but possibly innacurate predictions.  The method of k-nearest neighbors makes very mild structural assumptions: its predictions are often more accurate but can be unstable.

<h3>Linear Model and Least Squares</h3>
Given a vector of inputs,

$X^T = (X_1, X_2, ..., X_p)$

We predict the output $Y$ via the model

$ \hat{Y} = \hat{\beta_0} + \sum \limits_{j=1}^p X_j\hat{\beta_j}$

The term $\beta_0$ is the intercept, also known as the Bias in machine learning.  Often it is convenient to include the constant variable 1 in X, include $\beta_0$ in the vector of coefficients $\hat{\beta}$, and write the linear model in vector form

$\hat{Y} = X^T\hat{B}$

Viewed as a function across a p-dimensional input space, $f(X) =X^T\beta$ is linear, and the gradient $f'(X)=\beta$ is a vector in input space that points in the steepest uphill direction. 

The most common way to find a linear model is the method of leas squares.  We pick the coefficients $\beta$ to minimize the residual sum of squares,

$RSS(\beta) = \sum \limits_{i=1}^N (y_i - x_i^T\beta)^2$

$RSS(\beta)$ is a quadradic function of the parameters, and hence its minimum always exists, but may not be unique (multicolinearity).  The solution is easiest to characterize in matrix notation.

$RSS(\beta) = (y - X\beta)^T(y - X\beta)$

Where X is an N x p matrix with each row an input vector and y is an N-vector of the outputs in the training set.  Differentiating w.r.t. $\beta$ we get the normal equations,

$X^T(y - X\beta) = 0$

if $X^TX$ is nonsingular, then the unique solution is given by

$\hat{\beta} = (X^TX)^{-1}X^Ty$

The fitted value at the ith input $x_i$ is $\hat{y_i} = \hat{y_i}(x_i) = x_i^T \hat{\beta}$.  The entire fitted surface is characterized by the p parameters $\hat{\beta}$.  Intuitively, it seems that we do not need a very large data set to fit such a model.

We can use a linear model for classification with a simple rule, such as class 1 if $\hat{Y} > 0.5$, class 2 if $\hat{Y} \leq 0.5$.  Then our classification boundary is defined by the line $0.5 = \beta_0 + \beta_1x_1 + \beta_2x_2$

```python
#Fit Regression
dataset = np.matrix(np.concatenate((red, blue), axis=0))
X = dataset[0::, 1::]
y = dataset[0::, 0]
B = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)

# Decision Boundary is classified by XT.B = 0.5
classifier = lambda x: float((0.5 - B[0] - B[1]*x) / B[2])
domain = np.linspace(-2, 5, 100)
ranges = map(classifier, domain)

# Plot Results
plt.title("Linear Classification Boundary in 2-Dimensions")
plt.scatter(red[0::, 2], red[0::, 3], color=current_palette[2])
plt.scatter(blue[0::, 2], blue[0::, 3], color=current_palette[0])
plt.plot(domain, ranges, "--", color="grey")
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.ylabel("X2")
plt.xlabel("X1")
```

<h2> Nearest-Neighbor Methods</h2>
Nearest neighbor methods use those observations in the training set $\tau$ closest in input space to x to form $\hat{Y}$.  Specifically, the k-nearest neightbor for $\hat{Y}$ is defined as

$\hat{Y}(x) = \frac{1}{k} \sum \limits_{x_i \in N_{k(x)}} y_i$

Closeness implies a metric, a simple choice is Euclidean distance.  So, find the k closest observations to x in input space, and average their responses.  If the proportion of Y's is greater than 0.5, classify this point as Red, else classify it as Blue.  Regions should be more irregular, and correspond to local clusters where one class dominates.

```python
# Using 15 Nearest Neighbors
distance = lambda x, y: np.sqrt((x[0] - y[0])**2 + (x[1] - y[1])**2)
y = np.array(dataset[0::, 0])
X = np.array(dataset[0::, 2::])

# This is a poor way to do KNN, but i wanted to do it by hand
# Create new points for classification
domain = np.array([[p1, p2] for p1 in np.linspace(-2, 5, 50) for p2 in np.linspace(-2, 5, 50)])
classes = np.zeros(len(domain), np.float)
for i, point in enumerate(domain):
    distances = np.zeros(len(X), np.int)
    for j, c_point in enumerate(X):
        distances[j] = distance(c_point, point)
    classes[i] = np.round(np.mean(y[np.argsort(distances)[0:15]]))
    
# Plotting classification boundary
plt.title("15 Nearest Neighbor Classification")
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.scatter(domain[0::, 0], domain[0::, 1], color=map(color, classes), alpha=0.5, s=5)
plt.scatter(red[0::, 2], red[0::, 3], color=current_palette[2])
plt.scatter(blue[0::, 2], blue[0::, 3], color=current_palette[0])
```

```python
# Using 2 Nearest Neighbors
distance = lambda x, y: np.sqrt((x[0] - y[0])**2 + (x[1] - y[1])**2)
y = np.array(dataset[0::, 0])
X = np.array(dataset[0::, 2::])

# This is a poor way to do KNN, but i wanted to do it by hand
# Create new points for classification
domain = np.array([[p1, p2] for p1 in np.linspace(-2, 5, 50) for p2 in np.linspace(-2, 5, 50)])
classes = np.zeros(len(domain), np.float)
for i, point in enumerate(domain):
    distances = np.zeros(len(X), np.int)
    for j, c_point in enumerate(X):
        distances[j] = distance(c_point, point)
    classes[i] = np.round(np.mean(y[np.argsort(distances)[0:2]]))
    
# Plotting classification boundary
plt.title("2 Nearest Neighbor Classification")
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.scatter(domain[0::, 0], domain[0::, 1], color=map(color, classes), alpha=0.9, s=50)
plt.ylabel("X2")
plt.xlabel("X1")
```

The method K-nearest-neighbor averaging is defined exactly the same way for regression, although $k = 1$ would be an unlikely choice.  In terms of evaluating the error of a KNN model, the error on the training data is generally an increasing function of k.  For example, $k = 1$ misclassifies 0 points on training data, but this is not a very useful result.

The linear decision boundary from least squares is very smooth and stable to fit, though it relies heavily on the parametric model being somehat correct - low variance and high bias.  Alternatively, KNN procedures do not appear to rely on any stringent assumptions on parameters and can adapt to any situation.  Any partcular subregion, however, depends on a handfull of input points and their particular positions, and is thus wiggly and unstable - high variance and low bias.

<h1>The Curse of Dimensionality</h1>
With KNN methods, it seems as if with enough data we could always approximate the theoretically optimal conditional expectation, since we should be able to find a failry large neighborhood of observations of x and average them.  Our intuition and this approach breaks down in higher dimensionas, and the phenomenon is comonly refered to as the curse of dimensionality.

Consider the nearest-neighbor procedure for inputs uniformly distributed in a p-dimensional unit hypercube.  Suppose we send out a hypercubical neighborhood about a target point with the goal of capturing a fraction r of the observations.  Since this corresponds to a fraction r of the unit volume, the expected edge length will be $e_p(r) = r^{\frac{1}{p}}$.

For example,
In 10 dimensions $e_10(0.01) = 0.63$ and $e_10(0.1) = 0.80$,  While the entire range for each input is only 1.0.  So to capture $1\%$ or $10\%$ of the data, we must cover $63\%$ or $80\%$ of the range of each input variable.  Such neighborhoods are no longer "local".  

```python
df = pd.DataFrame([[p, r, r**(1 / p)] for p in [1, 2, 3, 10] for r in np.linspace(0, 1, 1000)])
df.columns = ['Dimensions','Fraction','Distance']
sns.lmplot("Fraction","Distance", data=df, hue='Dimensions', fit_reg=False)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title("The Curse of Dimensionality")
plt.xlabel("Fraction of Data Input")
plt.ylabel("Fraction of Cube Volume")
```

Another consequence of the sparse sampling in high dimensions is that all sample points are close to an edge of the sample.  Consider N data points uniformly distributed in a p-dimensional unit ball centered at the origin.  Suppose we consider a nearest-neighbor estimate at the origin.  The median distance from the origin to the closest data point is given by the expression

$d(p, N) = (1 - \frac{1}{2}^{\frac{1}{N}})^{\frac{1}{p}}$

A more complicated expression exists for the mean distance to the closest point.  For N = 500, p = 10, d(p, N)=0.52, more than halfway to the boundary.  Hence most data points are closer to the boundary of the sample space than to any other point.  This presents a problem for prediction becuase one must extrapolate from neighboring sample points rather than interpolate between them.

Another manifestation of the curse is that the sampling density is proportional to $N^\frac{-1}{p}$, where p is the dimension of the input space and N is the sample size.  Thus, if 100 represents a sense sample for a single input problem, $100^{10}$ is the sample size required for the same density with 10 inputs.  Thus in high dimensions all feasible training samples sparesly populate the input space.

<h1>K-Means Clustering - Not Supervised Learning!</h1>
The K-means algorithm puts N data points in an I-dimensional space into K clusters.  Each cluster is parameterized by a vector $m^k$ called its mean.  This requires a metric which defines the distance between two points.  The algorithm is an iterative two-step algorithm, first assign each data point to the closest cluster, then adjust the means to better fit the data points the are responsible for.  The algorithm has converged when assignment is unchanged

```python
k = np.random.normal(3,1,(2, 2)) # each row is the location of each cluster
X = np.array(dataset[0::, 2::])
distance = lambda x, y: np.sqrt(np.sum(np.square(x - y), axis=1))

for i in xrange(100):
    assignments = np.zeros((len(X), len(k)),np.bool)
    # assign points to clusters
    for j, point in enumerate(X):
        assignments[j, np.argmin(distance(point, k))] = True
    
    # update cluster means
    for j, cluster in enumerate(k):
        k[j] = np.mean(X[assignments[0::, j]], axis=0)
```

With just two means, we can draw a linear descision boundary by drawing a line orthagonal to the line that connects them at the halfway point.  Note that this is not based on whether or not the classification was correct, just optimal mean position

```python
point = np.mean(k, axis=0)
slope = 1 / (k[1][1] - k[0][1] / k[1][0] - k[0][0])
line = lambda x: slope*x - slope * point[0] + point[1]
domain = np.linspace(-2, 5, 100)
ranges = map(line, domain)
plt.title("K-Means unsupervised classification")
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.scatter(red[0::, 2], red[0::, 3], color=current_palette[2])
plt.scatter(blue[0::, 2], blue[0::, 3], color=current_palette[0])
plt.scatter(k[0][0], k[0][1], color="black",s=200, alpha=0.7)
plt.scatter(k[1][0], k[1][1], color="black",s=200, alpha=0.7)
plt.plot(domain, ranges, "--", color=current_palette[1])
```
