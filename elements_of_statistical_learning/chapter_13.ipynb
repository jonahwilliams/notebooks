```python
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline

current_palette = sns.color_palette()
sns.set_style('darkgrid')

# Mixture of Gaussian Dataset
red = np.array([[1, 1, np.random.normal(2, 1), np.random.normal(2, 1)] for i in range(100)])
blue = np.array([[0, 1, np.random.normal(0, 1), np.random.normal(0, 1)] for i in range(100)])
green = np.array([[2, 1, np.random.normal(3, 1), np.random.normal(0, 1)] for i in range(100)])
dataset = np.concatenate((red, blue, green), axis=0)
X = dataset[0::, 1::]
y = dataset[0::, 0]
```

<h1>Prototype Methods and Nearest Neighbors</h1>
Highly unstructured models which, while useless for interpretation, function very well as black box algorithms.

Given that our data consists of N pairs $(x_1, g_1), ..., (x_N,g_N)$ where $g_i$ is a class label taking values in $\{1, 2,...,K\}$.  Prototype methods represent the training data by a set of points in feature space, which are typically not training data.  i.e. we create different prototypical groups and assign data to some group based on a distance metric, for example.  Euclidean distance is appropriate for standardized qualitative features.  High dimensional problems and quantitative features require different considerations.

<h1>K-means Clustering</h1>
K-means clustering is a method for finding clusters and cluster centers in a set of unlabeled data.  It takes K, the number of cluster centers as a parameter.

To use K-means for labeled data, the algorithm is modified

  1. apply K-means clustering to the training data in each class separately, using R prototypes per class
  2. assign a class label to each of the $K x R$ prototypes
  3. classify a new feature $x$ to the class of the closest prototype

<h3>Learning Vector Quantization - LVQ</h3>

  1. Choose R inital prototypes for each class: $m_1(k), m_2(k), ..., m_R(k)$, k=1, 2, ...,K, for example, by sampling R training points at random from each class
  2. Sample a training point $x_i$ randomly(with replacement), and let $(j,k)$ index the closest prototype $m_j(k)$ to $x_i$.
  
       (a) If $g_i = k$ (i.e., they are in the same class), move the prototype towards the training point:
  
       $m_j(k) \leftarrow m_j(k) + \epsilon(x_i - m_j(k))$, where $\epsilon$ is the learning rate
       
       (b) If $g_i \neq k$ (i.e., they are in different classes), move the prototype away from the training point:
       
       $m_j(k) \leftarrow m_j(k) - \epsilon(x_i - m_j(k))$.
  
  3.  Repeat step 2, decreasing the learning rate with each iteration towards 0.
