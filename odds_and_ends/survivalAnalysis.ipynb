```python
from json import loads, dumps
from pyspark.sql import *
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import lifelines as ln
%matplotlib inline
sqlContext = HiveContext(sc)
```

```python
import datetime as dt
def survival_time(strt, mod, survived):
    # If a team is still active, it survives to the current day, else last modified date
    import datetime as dt
    modified = dt.datetime.fromtimestamp(mod / 1000)
    start = dt.datetime.fromtimestamp(strt / 1000)
    today = dt.datetime.today()
    if survived:
        return (today - start).days
    else:
        return (modified - start).days
    
sqlContext.registerFunction("time_fix", lambda x: dt.datetime.strftime(dt.datetime.fromtimestamp(x/1000),"%Y-%m-%d"))
sqlContext.registerFunction("survival_time", survival_time, IntegerType())
```

```python
schools = sqlContext.sql(
"""
SELECT time_fix(datecreated) as time_start, isactive as survived,
time_fix(datemodified) as time_end, survival_time(datecreated, datemodified, isactive) as delta,
state FROM Schools 
""")
schools.take(10)
```

```python
survival_df = pd.DataFrame(
    map(lambda x: (int(x['survived']),int(x['delta']), x['state']),
        map(lambda x: x.asDict(), schools.collect()))
)
```

```python
survival_df.columns = ['survived','time','state']
survival_df.head(10)
```

<h1>Basic Survival Analysis</h1>
Population remaining after given time

```python
from lifelines import KaplanMeierFitter
kmf = KaplanMeierFitter()
kmf.fit(survival_df['time'].values, event_observed=survival_df['survived'].values)
```

```python
kmf.plot()
plt.title("Survival Function")
plt.xlim(0, 2500)
```

```python
groups = survival_df['state']
ix = (groups == 'NE')

kmf.fit(survival_df['time'][ix], survival_df['survived'][ix], label='NE')
ax = kmf.plot()

kmf.fit(survival_df['time'][~ix], survival_df['survived'][~ix], label='Not NE')
kmf.plot(ax=ax)
plt.title("Survival Function")
plt.xlim(0, 2500)
```

```python
groups = survival_df['state']
ix1= (groups == 'NE')
ix2= (groups == 'CA')
ix3= (groups == 'NY')
ix4= (groups == 'CO')

kmf.fit(survival_df['time'][ix1], survival_df['survived'][ix1], label='NE')
ax = kmf.plot()

kmf.fit(survival_df['time'][ix2], survival_df['survived'][ix2], label='CA')
kmf.plot(ax=ax)

kmf.fit(survival_df['time'][ix3], survival_df['survived'][ix3], label='NY')
kmf.plot(ax=ax)

kmf.fit(survival_df['time'][ix4], survival_df['survived'][ix4], label='CO')
kmf.plot(ax=ax)

plt.title("Survival Function")
plt.xlim(0, 2500)
```

<h1>Estimating Hazard Curves</h1>
Below we estimate the cumulative hazard curve

```python
from lifelines import NelsonAalenFitter
T = survival_df["time"]
C = survival_df["survived"]


naf = NelsonAalenFitter()
naf.fit(T,event_observed=C)
naf.plot()
plt.xlim(0, 2500)
```

```python
groups = survival_df['state']
ix = (groups == 'NE')
T = survival_df["time"]
C = survival_df["survived"]


naf.fit(T[ix], event_observed=C[ix], label="Nebraska Teams")
ax = naf.plot()

naf.fit(T[~ix], event_observed=C[~ix], label="Non-Nebraska Teams")
naf.plot(ax=ax)
plt.xlim(0, 2500)
```
