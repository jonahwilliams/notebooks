<h2>Exploring the Data</h2>

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from ggplot import *
%matplotlib inline
```

```python
df = pd.read_csv('contact_counts.csv')
df['date'] = pd.to_datetime(df['date'])
df.index = df['date']
df = df[['call','email','date']]
df.head()
```

<p>With just a bit of munging, we have our data in an easily explorable format.  The Pandas datetime index allows one to easily subset by dates, and extract features such as day of the week.  Now we can go ahead and take a look at our time series data.  For the rest of this blog, I will only be refering to the email data of simplicity's sake.</p> 

```python
df = df[['email','date']]
ggplot(aes(x='date',y='email'), data=df) + \
    geom_line() + \
    xlab("Date") + \
    ylab("Volume") + \
    theme_bw() + \
    scale_x_discrete()
    
```

<p>The are very clearly some seasonal patterns in the data that can be exploited.  We're going to make a quick switch to the natural logarithm of the email counts; we really only care about the volume of emails in relative amounts.  As an added bonus, this should flatten out the upward trend in email volume as well</p>

```python
df['email'] = np.log(df['email'] + 1)
ggplot(aes(x='date',y='email'), data=df) + \
    geom_line() + \
    xlab("Date") + \
    ylab("Ln(Volume)") + \
    theme_bw() + \
    scale_x_discrete()
```

<p> As we can see from above, weekly patterns are a lot noiseier than the yearly pattern.  It would extremely tedious to explore all the different facets of the overall pattern. <h2>Learning Patterns in Time-Series</h2>
<p>Since we are attempting to predict a real valued output, why not start with a simple linear regression?   Our equation would be of the form
\begin{align}  
Ln(Y_i) = \beta_0 + \beta_1Ln(X_1) + ... + \beta_{n}Ln(X_{n})
\end{align}
Where each term is the logarithm of the previous day's volume, and the target is logarithm the next day's volume.  Treating this problem as an autoregressive linear regression has several benefits:  (1) It has a global optimum solution, and can be solved in closed form (2) It has nice analytical properties we can use to measure variance.</p>
<p>First, we will generate the necessary features, 380 days of lagged email counts.  Then, in order to evaluate the effectiveness of our model we will split the dataset into two parts, a test set and a training set.  The model will be fit on the training set and then evaluated on the test set.  We will use sklearn's linear regression implementation</p>

```python
df = df[['email']]
for i in xrange(1, 380):
    term = 'n-' + str(i)
    df[term] = df['email'].shift(i)
df = df.dropna(axis=0)
training_data = df.ix['2013'].values
testing_data = df.ix['2014'].values
clf = linear_model.LinearRegression()
```

```python
clf.fit(training_data[0::, 1::], training_data[0::, 0])
train = clf.predict(training_data[0::, 1::])
train = pd.DataFrame([[i, np.exp(train[i]), 'Predicted'] for i in xrange(len(train))])
results = pd.DataFrame([[i, np.exp(training_data[i, 0]), 'Actual'] for i in xrange(len(train))])
data = pd.concat([train, results], axis=0)
data.columns = ['i','value','variable']
ggplot(aes(x='i', y='value', group='variable', color='variable'), data=data) + \
    geom_line() + \
    scale_x_continuous() + \
    xlim(low=0,high=300) + \
    ylim(low=0, high=2000) + \
    xlab("") + \
    ylab("Volume") + \
    ggtitle("In Sample Prediction")
```

```python
test = clf.predict(testing_data[0::, 1::])
test = pd.DataFrame([[i, np.exp(test[i]), 'Predicted'] for i in xrange(len(test))])
results = pd.DataFrame([[i, np.exp(testing_data[i, 0]), 'Actual'] for i in xrange(len(test))])
data = pd.concat([test, results], axis=0)
data.columns = ['i','value','variable']
ggplot(aes(x='i', y='value', group='variable', color='variable'), data=data) + \
    geom_line() + \
    scale_x_continuous() + \
    xlim(low=0,high=300) + \
    ylim(low=0, high=2000) + \
    xlab("") + \
    ylab("Volume") + \
    ggtitle("Out of Sample Prediction")
```

<p>As we can see from above, the test set fit is extremely poor.  Since we didn't know which autoregressive (lagged volume) terms were important, we included all of them.  With so many parameters, however, a simple linear model will almost certainly overfit.  We could manually check every different combination of autoregressive terms, perform t-tests of significance and....yeah no one wants to do that.</p>
<h2>Automatic Variable Selection With Lasso</h2>
<p>Consider a linear regression as above, but subject to the following constraint:</p>
\begin{align}
\sum \limits_{i=1}^n \mid\beta_i\mid \leq c
\end{align}
<p>The constant c allows us to restrict how many variables we allow into the model, and it follows that if c is large enough we get the same result as a simple linear regression.  If c is restrictive enough, variables with low predictive power will be completely excluded from the model.  Additionally, we will not reach a global optimum so numerical methods must be used to solve the regression. A nice way to this is with Least Angle Regression:  For a given step size, add a part of the parameter most correlated with the residuals to the model until we reach c.</p>
<p>Below we use python's implementation to fit a model to our training set, and evaluate on our test set.  sklearn's Lasso takes $\alpha$ as a parameter instead of a constant c, and instead minimizes.</p>
\begin{align}
(1 / (2 * n_{samples})) * ||y - Xw||^2_2 + alpha * ||w||_1
\end{align}
This is equivalent to the above constraint.

```python
clf = linear_model.Lasso(alpha=0.01)
clf.fit(training_data[0::, 1::], training_data[0::, 0])
train = clf.predict(training_data[0::, 1::])
train = pd.DataFrame([[i, np.exp(train[i]), 'Predicted'] for i in xrange(len(train))])
results = pd.DataFrame([[i, np.exp(training_data[i, 0]), 'Actual'] for i in xrange(len(train))])
data = pd.concat([train, results], axis=0)
data.columns = ['i','value','variable']
ggplot(aes(x='i', y='value', group='variable', color='variable'), data=data) + \
    geom_line() + \
    scale_x_continuous() + \
    xlim(low=0,high=300) + \
    ylim(low=0, high=2000) + \
    xlab("") + \
    ylab("Volume") + \
    ggtitle("In Sample Prediction")
```

```python
test = clf.predict(testing_data[0::, 1::])
test = pd.DataFrame([[i, np.exp(test[i]), 'Predicted'] for i in xrange(len(test))])
results = pd.DataFrame([[i, np.exp(testing_data[i, 0]), 'Actual'] for i in xrange(len(test))])
data = pd.concat([test, results], axis=0)
data.columns = ['i','value','variable']
ggplot(aes(x='i', y='value', group='variable', color='variable'), data=data) + \
    geom_line() + \
    scale_x_continuous() + \
    xlim(low=0,high=300) + \
    ylim(low=0, high=2000) + \
    xlab("") + \
    ylab("Volume") + \
    ggtitle("Out of Sample Prediction")
```

<p>The model fits our test data much better than a simple linear regression;
by excluding uncorrelated parameters we prevent some overfitting.  It can be seen below that most of the model's coefficients are forced towards zero, while the autoregressive terms with high predictive power enter the model.  Though we arbitrarily selected our $\alpha$, one could use cross validation to choose $\alpha$ as well.</p>
